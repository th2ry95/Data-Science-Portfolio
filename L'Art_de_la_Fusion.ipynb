#Générer tes notre sources de données

```python
#Importation de la bibliothèque
import pandas as pd

# Fichier 1 : Les transactions (Ventes)
ventes_data = {
    'Transaction_ID': [101, 102, 103, 104, 105],
    'Product_ID': ['P001', 'P003', 'P001', 'P002', 'P003'],
    'Quantite': [2, 1, 5, 2, 3]
}
df_ventes = pd.DataFrame(ventes_data)

# Fichier 2 : Le catalogue (Produits)
produits_data = {
    'Product_ID': ['P001', 'P002', 'P003'],
    'Nom_Produit': ['Clavier Sans Fil', 'Souris Gaming', 'Écran 4K'],
    'Prix_Unitaire': [45, 25, 300],
    'Categorie': ['Accessoires', 'Accessoires', 'Écrans']
}
df_catalogue = pd.DataFrame(produits_data)

print("Les deux fichiers sont prêts à être fusionnés !")
```

```python
#Affichons la base de donné(ventes_data)

df_ventes.head()
```

```python
#Affichons la base de donné(produits_data)
df_catalogue.head()
```

#Comment créer un seul grand tableau qui contient TOUTES les infos (Ventes + Catalogue) grâce à leur point commun (Product_ID) ?

```python
# On fusionne nos deux base de données pour n'en faire qu'une

merged_df = df_ventes.merge(df_catalogue)
merged_df
```

# Trouvons le CA de l'entreprise

```python
# On va calculer le CA suite à notre calcul suivant Pu * Qa

merged_df['Total_vente'] = merged_df['Prix_Unitaire'] * merged_df['Quantite']
```

```python
# Observons le résultat
merged_df
```

# Terminons avec une analyse croisé en foctions de la catégorie par rapport au chiffre d'affaire pour voir la plus rentable

```python
merged_df.groupby('Categorie')['Total_vente'].sum().sort_values(ascending=False)
```

 Le calcul du chiffre d'affaires total par catégorie révèle que la catégorie 'Écrans' est la plus performante, générant un total de 1200. En comparaison, la catégorie 'Accessoires' a contribué à hauteur de 365 au chiffre d'affaires. Ces résultats indiquent clairement que les 'Écrans' représentent la catégorie de produits la plus rentable pour l'entreprise.
